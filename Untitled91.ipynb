identifying the null values and duplicate values

```python
def identify_null_values(df):
  null_values = df.apply(lambda x:x.isnull().sum())    #calculating the total number of null columns
  return null_values

def identify_duplicate_values(df):
  duplicate_values = df.apply(lambda x:x.duplicated().sum())  #return the column consist any duplicate
  return duplicate_values

#returning the columns with count of null values and duplicate values
null_values = identify_null_values(churn_df)
duplicate_values = identify_duplicate_values(churn_df)

#returning the total percentage of null values in columns
percentage_null_columns = (null_values / len(churn_df)) * 100

#returning the total percentage of duplicate values in columns
percentage_duplicate_columns = (duplicate_values / len(churn_df)) * 100

print(f" Percentage of null values in columns : {percentage_null_columns}")
print(f" Percentage of duplicate values in columns : {percentage_duplicate_columns}")
```

# Splitting the features into training and testing for faster processing

```python
X = churn_df.drop(['Churn'],axis=1)
y = churn_df['Churn']
```

```python
class FeatureSplitter:

  def __init__(self,X,y):
    self.X = X      #initializing the input features
    self.y = y      #initializing the output features
    self.train_size = 0.7   #initializing the train size with 70%
    self.test_size = 0.3    #initializing the test size with 30%
    self.random_state = 42   #initializing the random state as 42 for data consistancy
    self.shuffle = True

  def split_features(self):
    try:
      X_train , X_test , y_train , y_test = train_test_split(
          self.X,self.y,
          train_size = self.train_size,
          test_size = self.test_size,
          random_state = self.random_state,
          shuffle = self.shuffle)
      return X_train , X_test , y_train , y_test
    except Exception as e:
      raise e


#calling the class with object
feature_splitter = FeatureSplitter(X,y)

#splitting the features into training and testing
X_train , X_test , y_train , y_test = feature_splitter.split_features()

print(f" training features : {X_train.shape}")
print(f" testing features : {X_test.shape}")
```

# Automating Feature selection Using Decision Tree Algorithm

```python
#creating an pipeline interface with some preprocessing workflow
preprocessing_workflow = (['StandardScaler',StandardScaler()],['model',DecisionTreeClassifier(random_state=42)])

def create_pipeline_workflow(preprocessing_workflow):
  try:
    feature_automation_pipeline = Pipeline(preprocessing_workflow)
    return feature_automation_pipeline
  except Exception as e:
    raise e

#calling the function
feature_automation_pipeline = create_pipeline_workflow(preprocessing_workflow)
feature_automation_pipeline
```

```python
#creating an hyperparamter tuning
hyperparameter_tuning = {
    'model__max_depth':[3,4,5],
    'model__min_samples_split':[2,3,4],
    'model__min_samples_leaf':[1,2,3]
}

grid_search = GridSearchCV(feature_automation_pipeline,hyperparameter_tuning,cv=5,n_jobs=-1)
model = grid_search.fit(X_train,y_train)
print(f" Decision Tree model : {model}")
```

# Building Churn model that will impact high perforance on churn modelling

1. Logsistic Regression

```python
class Logistic_Regression:

  def __init__(self, X_train, y_train, X_test, y_test):
    self.X_train = X_train      # initializing the input features
    self.y_train = y_train      # initializing the output features
    self.X_test = X_test        # initializing the test input features
    self.y_test = y_test        # initializing the test output features
    self.model = LogisticRegression(random_state=42) # Initialize the model instance
    self.preprocessing_workflow = [('StandardScaler', StandardScaler()), ('model', self.model)] # Define the pipeline structure
    self.hyperparameter = {'model__C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
                           'model__penalty': ['l1', 'l2'],
                           'model__solver': ['liblinear', 'saga']} # Define hyperparameters


  def create_logistic_regression_pipeline(self):
    """ returning the pipeline for logistic regression """
    try:
      logistic_regression_pipeline = Pipeline(self.preprocessing_workflow)
      return logistic_regression_pipeline
    except Exception as e:
      raise e

  def build_model(self, logistic_regression_pipeline):
    """ Builds and fits the GridSearchCV model """
    try:
      grid_search = GridSearchCV(logistic_regression_pipeline, self.hyperparameter, cv=5, n_jobs=-1)
      model = grid_search.fit(self.X_train, self.y_train)
      return model
    except Exception as e:
      raise e

# calling the class with object
lr = Logistic_Regression(X_train, y_train, X_test, y_test)

# Call the methods on the instance
logistic_regression_pipeline = lr.create_logistic_regression_pipeline()
model = lr.build_model(logistic_regression_pipeline)

print(f" Logistic Regression model : {model}")
```

```python
#predicting the model with probability of churn
churn_pred = model.predict(X_test)

#predicting the model with probability
churn_pred_proba = model.predict_proba(X_test)[:,1]
print(f" Logistic regression prediction : {churn_pred}")
print(f" Logistic regression prediction probability : {churn_pred_proba}")

df = {'churn_result':churn_pred,'churn_probability':churn_pred_proba}
df = pd.DataFrame(df)


sns.barplot(df,x='churn_result',y='churn_probability')
plt.xlabel('Churn Result')
plt.ylabel('Churn Probability')
plt.title('Churn Probability Distribution')
plt.show()
```

Optimize the Result using Random Forest Algorithm

```python
class Random_Forest:
  def __init__(self,X_train,X_test,y_train , y_test):
    self.X_train = X_train
    self.X_test = X_test
    self.y_train = y_train
    self.y_test = y_test
    self.model = RandomForestClassifier(random_state=42)
    self.preprocessing_workflow = [('StandardScaler',StandardScaler()),('model',self.model)]
    self.hyperparameter = {'model__n_estimators':[100,200,300],
                           'model__max_depth':[None,5,10],
                           'model__min_samples_split':[2,5,10],
                           'model__min_samples_leaf':[1,2,4]}

  def create_random_forest_pipeline(self):
    try:
      random_forest_pipeline = Pipeline(self.preprocessing_workflow)
      return random_forest_pipeline
    except Exception as e:
      raise

  def build_model_random_forest(self,random_forest_pipeline):
    try:
      grid_search = GridSearchCV(random_forest_pipeline,self.hyperparameter,cv=5,n_jobs=-1)
      model = grid_search.fit(self.X_train,self.y_train)
      return model
    except Exception as e:
      raise


#calling the instance
rf = Random_Forest(X_train,X_test,y_train,y_test)

#calling the methods
random_forest_pipeline = rf.create_random_forest_pipeline()
rf_model = rf.build_model_random_forest(random_forest_pipeline)

print(f" Random Forest model : {rf_model}")
```

```python
#implementing XGBoosting classifier
class XGBoost:

  def __init__(self,X_train,y_train,X_test,y_test):
    self.X_train = X_train
    self.y_train = y_train
    self.X_test = X_test
    self.y_test = y_test
    self.model = XGBClassifier(random_state=42,objective='binary:logistic')  #uses for sigmoid classification
    self.preprocessing_workflow = [('StandardScaler',StandardScaler()),('model',self.model)]
    self.hyperparameter_tuning = {'model__n_estimators':[100,200,300],
                                  'model__max_depth':[3,4,5],
                                  'model__learning_rate':[0.01,0.1,0.2]}

  def create_boosting_pipeline(self):
    """ Return XGBoost Pipeline with preprocessing and model """
    try:
      boosting_pipeline = Pipeline(self.preprocessing_workflow)
      return boosting_pipeline
    except Exception as e:
      raise e

  def create_xgbboost_model(self,boosting_pipeline):
    """ Builds and fits the GridSearchCV model """
    try:
      grid_search = GridSearchCV(boosting_pipeline,self.hyperparameter_tuning,cv=5,n_jobs=-1)
      model = grid_search.fit(self.X_train,self.y_train)
      return model
    except Exception as e:
      raise e


#calling the instance of XGBoost Model
xgb = XGBoost(X_train,y_train,X_test,y_test)

boosting_pipeline = xgb.create_boosting_pipeline()

xgb_model = xgb.create_xgbboost_model(boosting_pipeline)

print(f" XGBoost model : {xgb_model}")
```

```python
class Multi_layer_neuron:

  def __init__(self,X_train_copy,y_train,X_test_copy,y_test):
    self.X_train = X_train_copy
    self.y_train = y_train
    self.X_test = X_test_copy
    self.y_test = y_test
    self.model = Sequential()  #using sequential model
    self.optimizers = Adam(learning_rate=0.001)
    self.loss = BinaryCrossentropy()  #using binary loss function to equalize the distrbution of 0 and 1
    self.metrics = ['accuracy']
    self.callbacks = EarlyStopping(monitor='val_loss',patience=10, restore_best_weights=True) #monitor the training loss error for over epochs 1000
    self.epochs = 1000
    self.input_dim = X_train.shape[1]
    self.Dropout = Dropout(0.3)  #using dropouts for overfitting prone

  def train_Multi_layer_neuron(self):
    try:
      """ Training 3 layers of Neurons with relu activation function """
      # First layer
      self.model.add(Dense(units=128,activation='relu',input_dim = self.input_dim))
      self.model.add(BatchNormalization())
      self.model.add(self.Dropout)

      #second layer
      self.model.add(Dense(units=64,activation='relu'))
      self.model.add(BatchNormalization())
      self.model.add(self.Dropout)

      #third layer
      self.model.add(Dense(units=32,activation='relu'))
      self.model.add(BatchNormalization())  #speed up the process of learning
      self.model.add(self.Dropout)

      #Output layer which is feed forward neurons
      self.model.add(Dense(units=1,activation='sigmoid'))

      #adding optimizers and loss function to minimize the erros
      model_summary = self.model.compile(optimizer=self.optimizers,loss=self.loss,metrics=self.metrics)

      #train the model 1000 epochs loops and monitor the val_loss
      history = self.model.fit(self.X_train,self.y_train,epochs=self.epochs,callbacks=self.callbacks,validation_split=0.2)
      # Return the history object and the compiled model instance
      return history, self.model

    except Exception as e:
      raise e


#calling the instance of Multi layers neurons
MLN = Multi_layer_neuron(X_train_copy,y_train,X_test_copy,y_test)

#calling the method and unpacking the returned tuple
history, trained_model = MLN.train_Multi_layer_neuron()
```

```python
class Store_results:
  def __init__(self,X_train,y_train,X_test,y_test):
    self.X_train = X_train
    self.y_train = y_train
    self.X_test = X_test
    self.y_test = y_test
    self.prediction_model = trained_model  #saving the model for churn prediction
    self.probability_model = rf_model #saving the model for probability of risk

  def store_training_data(self):
    """ Storing the training data """
    try:
      self.X_train.to_csv('X_train.csv',index=False)
      self.y_train.to_csv('y_train.csv',index=False)
      self.y_train.to_csv('X_test.csv',index=False)
      self.y_test.to_csv('y_test.csv',index=False)
    except Exception as e:
      raise e

  def store_prediction_prob_model(self):
    """ Storing the prediction probability model """
    try:
      prediction_prob_model = joblib.dump(self.probability_model,'probability_model.pkl')
      return prediction_prob_model
    except Exception as e:
      raise e

  def store_prediction_model(self):
    """ Storing the prediction model """
    try:
      prediction_model = self.prediction_model.save('prediction_model.keras')
      return prediction_model
    except Exception as e:
      raise e


#callingt the above instance class
Store_results = Store_results(X_train,y_train,X_test,y_test)
Store_results.store_training_data()
Store_results.store_prediction_model()
Store_results.store_prediction_prob_model()
```
